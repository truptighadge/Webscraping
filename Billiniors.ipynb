```python
#Important libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd

#Fetching a given URL
url="https://en.wikipedia.org/wiki/List_of_Indians_by_net_worth"
page=requests.get(url)
soup=BeautifulSoup(page.content,'html5lib')
soup
```

```python
tdata=soup.find(class_="wikitable sortable")
tdata
```

```python
heads=[t.get_text().strip().replace('\n',' ') for t in tdata.find_all('th')]
heads
```

```python
#tdata.find_all('tr')[1:]
res=[[g.get_text().strip() for g in t.find_all('td')] for t in tdata.find_all('tr')[1:]]
res
```

```python
res2=[t.find_all('td')[2].find('img')['title'] for t in tdata.find_all('tr')[1:]]
res2
```

```python
len(res2),len(res)
```

```python
for i in range(len(res)):
    res[i][2]=res2[i]
    print(res[i])
```

```python
d={k:[] for k in heads}
for p in res:
    d['Rank'].append(p[0])
    d['Name'].append(p[1])
    d['Wealth Change'].append(p[2])
    d['Net worth (USD)'].append(p[3])
    d['Company'].append(p[4])
    d['Sources of wealth'].append(p[5])
df=pd.DataFrame(d)
df
```

```python
df.to_csv("Billiniors.csv",index=False)
```
