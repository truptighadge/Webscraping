```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url="http://www.delhisldc.org/Loaddata.aspx?mode=01/01/2022"
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.content,'html5lib')
soup
```

```python
data=soup.find(id="ContentPlaceHolder3_DGGridAv")
data
```

```python
tdata=data.find_all('tr')
tdata
```

```python
theads=[i.get_text() for i in tdata[0].find_all('td')]
theads
```

```python
tb=tdata[1:]
tb
```

```python
fab=tb[0]
fab
```

```python
firstbr=[j.get_text() for j in tb[0].find_all('td')]
firstbr
```

```python
w=[[j.get_text() for j in k.find_all('td')]for k in tdata[1:]]
w
```

```python
d={k:[] for k in theads}
for p in w:
    d['TIMESLOT'].append(p[0])
    d['DELHI'].append(p[1])
    d['BRPL'].append(p[2])
    d['BYPL'].append(p[3])
    d['NDPL'].append(p[4])
    d['NDMC'].append(p[5])
    d['MES'].append(p[6])
df=pd.DataFrame(d) 
df
```
